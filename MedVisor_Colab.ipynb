## 3. Configure the LLM API key

Get a free key from https://openrouter.ai/keys (OpenRouter has free models).

```python
import os

# ===== Set your API key here =====
os.environ["LLM_BASE_URL"] = "https://openrouter.ai/api/v1"
os.environ["LLM_API_KEY"]  = "sk-or-v1-XXXXXXXXXXXXXXXXXXXX"  # <-- paste your key
os.environ["LLM_MODEL"]    = "meta-llama/llama-3.1-8b-instruct"

# Image model: "sdxl-turbo" (fast) or "sd15"
os.environ["IMAGE_MODEL"] = "sdxl-turbo"

print("Config set.")

# --- Alternative: load from Colab Secrets ---
# from google.colab import userdata
# os.environ["LLM_API_KEY"] = userdata.get("LLM_API_KEY")
```
